```python
df['clean_text'] = hero.clean(df['headline_text'])
```

```python
df['tfidf'] = hero.tfidf(df['clean_text'],max_features=100)
```

```python
from sklearn.cluster import KMeans
vector = df['tfidf'].values.tolist()
```

```python
kmeans = kmeans = KMeans(
        n_clusters=5,
        init="k-means++",
        algorithm="auto"
    )
kmeans.fit(vector)
clusters = kmeans.predict(vector)
```

```python
np.unique(clusters)
```

```python
df['label'] = clusters
```

```python
df.head()
```

```python
df['pca'] = hero.pca(df['tfidf'])
```

```python
df.head()
```

```python
df = df.sample(frac=1)
hero.scatterplot(df.head(50000),col='pca',color = 'label',title="ABC NEWS Data")
```
